# Under sample majority class

```python
df_under_sample = pd.concat([df3_0 , count_1_df],axis = 0)
df_under_sample.shape
```

```python
df_under_sample.Churn.value_counts()
```

```python
X = df_under_sample.drop(['Churn'],axis = 'columns')
y = df_under_sample['Churn']
X.shape
```

```python
def ANN_churn(X_train , y_train , X_test , y_test , loss):
    
    model = keras.Sequential([
        keras.layers.Dense(26 , input_shape = (26,),activation = 'relu'),
        keras.layers.Dense(15 ,activation = 'relu'),
        keras.layers.Dense(1 ,activation = 'sigmoid')
    ])
    
    model.compile(optimizer = 'adam',loss = loss, metrics = ['accuracy'])
    
    model.fit(X_train , y_train , epochs = 100)
    
    print(model.evaluate(X_test , y_test))
    
    y_pred = model.predict(X_test)
    y_pred = np.round(y_pred)
    
    print('/////Classification  Report')
    print(classification_report(y_test , y_pred))
    
    return y_pred 
    
```

```python
from sklearn.model_selection import train_test_split 
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size= 0.2 , random_state = 0,stratify = y)
```

```python
y_train.value_counts()
```

```python
y_pred = ANN_churn(X_train , y_train , X_test , y_test, 'binary_crossentropy')
```

```python
X = df2.drop('Churn',axis = 'columns')
y = df2['Churn']
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy = 'minority')
X_sm , y_sm = smote.fit_sample(X, y )
y_sm.value_counts()
```

```python
X_train , X_test , y_train , y_test = train_test_split(X_sm , y_sm , test_size = 0.2 , random_state = 0 , stratify = y_sm)
```

```python
y_pred = ANN_churn(X_train , y_train , X_test , y_test, 'binary_crossentropy')
```
